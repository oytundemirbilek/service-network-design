# 1. Why Advanced Air Mobility (AAM) in NYC?

## 1.1 Rising Congestion & High Value of Time

- **Severe Ground Congestion**  
  Manhattan is one of the most congested urban centers in the world; average speeds during peak hours can be extremely low.

- **High Willingness-To-Pay (WTP)**  
  Many business travelers, tourists, and affluent residents value time savings significantly. Shortening a 60–90 minute car ride to a 10–15 minute urban flight can attract premium fares.


## 1.2 Technological Feasibility & Regulatory Momentum

- **eVTOL Advances**  
  Electric vertical takeoff and landing (eVTOL) technologies are rapidly maturing. Several companies (Joby, Archer, Wisk) have targeted NYC as a potential launch market.

- **Local and Federal Support**  
  NASA’s AAM program, FAA’s evolving eVTOL certification standards, and NYC’s history of helicopter travel corridors provide a springboard for near-future demonstration projects.

## 1.3 Why Service Network Design?

- **Where to Fly & How Often**  
  Early-stage operators must select which vertiports to open, which routes to serve, how frequently to fly, and how to price them.

- **Infrastructure Constraints**  
  Rooftop vertiports, existing heliports, or new barge-based landing pads each have capacity, regulatory, and cost implications.

- **Balancing Demand & Profit**  
  Not all routes will be profitable. Operators must weigh the revenue potential (passenger fares) against operational costs.

# 2. Mathematical Model for NYC AAM Service Network


## 2.1 Sets & Indices

- $I$: The set of potential vertiports (nodes)  *(Index: $i \in I$, or $j \in I$ for distinct origin/destination)*

- $A$: The set of directed arcs $(i,j)$ representing feasible aerial routes   *(Index: $a \in A$)*

- $S$: The set of service (aircraft) types or modes (e.g., different eVTOL models) *(Index: $s \in S$)*

- $P$: The set of passenger segments (e.g., Business, Tourist, VIP)  *(Index: $p \in P$)*

- $T$: The set of discrete time periods (e.g., morning peak, evening peak) *(Index: $t \in T$)*
<br>

## 2.2 Parameters

**Revenue & Costs**
- $r_{p,a}$: Revenue per unit of passenger type $p$ on arc $a$  
- $v_{p,a}$: Variable cost for transporting passenger type $p$ on arc $a$  
- $c_{s,a}$: Cost of operating service $s$ on arc $a$ (per flight or time period)  
- $f_s$: Fixed cost for activating service $s$ (e.g., daily overhead)

**Capacity & Demand**
- $u_{s,a}$: Capacity of service $s$ on arc $a$ (max passengers per flight/period)  
- $d_{p,t}$: Demand for passenger segment $p$ in time $t$  
- $M$: A large constant (used in frequency constraints)

**Feasibility & Accessibility**
- $\alpha_{p,a}$: Accessibility measure (flight time, distance, or noise) for passenger $p$ on arc $a$  
- $\beta$: Maximum allowable threshold (e.g., distance or time)

**Budget & Coverage**
- $B$: Available budget if needed (optional constraint)  
- $\kappa_s$: Minimum service coverage requirement for service $s$

<br>


## 2.3 Decision Variables

**Service Activation & Frequency**
- $y_{s,t} \in \{0,1\}$: Binary variable; 1 if service $s$ is operated at time $t$, else 0  
- $\nu_{s,t} \in \mathbb{N}_0$: Integer frequency of service $s$ at time $t$ (e.g., flights per time block)

**Service Capacity Allocation**
- $z_{s,a,t} \ge 0$: Volume of service $s$ used on arc $a$ at time $t$ (allocated seats or capacity)

**Passenger Flow**
- $x_{p,a,t} \ge 0$: Flow of passenger type $p$ on arc $a$ at time $t$ (e.g., passenger count)

> **Note:** If $\alpha_{p,a} > \beta$, then that route is not feasible for passenger type $p$. Thus, $x_{p,a,t} = 0$ in such cases.

<br>

## 2.4 Objective Function: Maximize Profit

$$
\max \Bigg(\underbrace{\sum_{t \in T} \sum_{p \in P} \sum_{a \in A} r_{p,a} \, x_{p,a,t}}_{\text{Total Revenue}}
\;-\;
\underbrace{\Big[\sum_{t \in T} \sum_{s \in S} f_s \, y_{s,t}
\;+\;
\sum_{t \in T} \sum_{p \in P} \sum_{a \in A} v_{p,a} \, x_{p,a,t}
\;+\;
\sum_{t \in T} \sum_{s \in S} \sum_{a \in A} c_{s,a} \, z_{s,a,t}\Big]}_{\text{Total Cost}}\Bigg)
$$

- **Total Revenue**: Summation of revenues from passenger flows on each arc/time.  
- **Total Cost**:  
  1. Fixed cost $f_s$ for each service $s$.  
  2. Variable cost $v_{p,a}$ for passenger flows $ x_{p,a,t}$.  
  3. Service-specific cost $ c_ {s,a} $ for allocated capacity $z_{s,a,t} $.

<br>

## 2.5 Constraints

**(1) Demand Satisfaction (if we must serve all demand)**

$$
\sum_{a \in A} x_{p,a,t} \;\ge\; d_{p,t},
\quad \forall p \in P,\;\forall t \in T.
$$

  - **Interpretation:** For each passenger type $p$ and each time period $t$, the total flow allocated across all arcs $a$ must be at least equal to the demand $d_{p,t}$. In other words, if 100 passengers in period $t$ need to travel, we must schedule enough capacity/flow so that 100 passengers can be served.


<br>

**(2) Flow Conservation (if multi-leg routes exist)**

$$
\sum_{a \in A^{\text{in}}(i)} x_{p,a,t}
=
\sum_{a \in A^{\text{out}}(i)} x_{p,a,t},
\quad \forall i \in I,\; \forall p \in P,\; \forall t \in T.
$$

   - **Interpretation:** The flow of passengers of type $p$ arriving at node $i$ in period $t$ must equal the flow departing node $i$ in period $t$. This ensures logical consistency—passengers do not accumulate or vanish at an intermediate node unless it is their final destination.


<br>

**(3) Service Selection**

$$
z_{s,a,t} \;\le\; u_{s,a} \cdot y_{s,t},
\quad \forall s \in S,\;\forall a \in A,\;\forall t \in T.
$$

   - **Interpretation:** If $y_{s,t} = 0$ (meaning service $s$ is not operated in period $t$), then $z_{s,a,t}$ (the capacity allocated on arc $a$) must be zero. Conversely, if $y_{s,t} = 1$, $z_{s,a,t}$ can be up to the maximum capacity $u_{s,a}$.


<br>

**(4) Frequency Constraint**

$$
\nu_{s,t} \cdot y_{s,t} \;\le\; M,
\quad \forall s \in S,\;\forall t \in T.
$$
   - **Interpretation:** If $y_{s,t} = 1$, then $\nu_{s,t}$ (the integer frequency of service $s$) can be up to $M$. If $y_{s,t} = 0$, then $\nu_{s,t} = 0$. This keeps frequency at zero if the service is not activated.


<br>

**(5) Capacity Constraint**

$$
\sum_{p \in P} x_{p,a,t}
\;\le\;
\sum_{s \in S} z_{s,a,t},
\quad \forall a \in A,\;\forall t \in T.
$$
   - **Interpretation:** The sum of all passenger flows $x_{p,a,t}$ (for all types $p$) on arc $a$ in period $t$ cannot exceed the sum of allocated capacities $z_{s,a,t}$ provided by each active service $s$.


<br>

**(6) Accessibility Constraint**

If $\alpha_{p,a} > \beta$, then:
$$
x_{p,a,t} = 0,
\quad \forall p \in P,\;\forall a \in A,\;\forall t \in T.
$$
   - **Interpretation:** If the route $a$ for passenger type $p$ exceeds the threshold $\beta$ (e.g., maximum distance or flight time), then no flow is permitted on that arc. It effectively “turns off” arcs that are beyond acceptable limits.


<br>

**(7) Budget Constraint (Optional)**

$$
\sum_{t \in T} \sum_{s \in S} f_s \, y_{s,t}
\;+\;
\sum_{t \in T} \sum_{p \in P} \sum_{a \in A} v_{p,a} \, x_{p,a,t}
\;\;\le\;\; B.
$$

   - **Interpretation:** The sum of fixed service costs ($f_s$) and variable flow costs ($v_{p,a} \cdot x_{p,a,t}$) must not exceed budget $B$. This constraint is vital when capital or operating funds are limited.


<br>

**(8) Minimum Service Coverage (If Required)**

$$
\sum_{t \in T} y_{s,t} \;\ge\; \kappa_s,
\quad \forall s \in S.
$$
   - **Interpretation:** The service $s$ must be activated at least $\kappa_s$ times over the entire horizon. This might reflect a policy requirement, a contractual obligation, or a basic “brand presence” expectation.

# 3. Data preparation for Manhattan AAM Case


Designing a **Service Network** to determine optimal vertiport locations in Manhattan (or the broader NYC area) requires **comprehensive datasets** that capture **mobility patterns**, **urban infrastructure**, **socioeconomic factors**, and **regulatory constraints**. Below, we map **each model component** (sets, indices, parameters, and decision variables) to **possible data sources** and **strategies** for collection and assembly.


## 3.1. Sets & Indices

### 3.1.1 \( I \): Potential Vertiports (Nodes)
- **Definition**: Each element \(i \in I\) represents a candidate site for a vertiport.  
- **Data Requirements**:  
  - **Geospatial Coordinates** (latitude/longitude, building rooftops, ground plots, existing heliports)  
  - **Zoning/Permits** (zoning code, land use type)  
  - **Potential Site Size** (area, height restrictions)  
- **Sources**:  
  - **NYC Planning Department**: Zoning maps, land use shapefiles.  
  - **NYC OpenData**: Building footprints, land parcels, current heliport locations.  
  - **Real Estate Databases** (e.g., CoStar, Zillow for commercial properties).

### 3.1.2 \( A \): Directed Arcs (Routes)
- **Definition**: Each arc \((i,j)\) in \(A\) represents a feasible flight corridor from vertiport \(i\) to vertiport \(j\).  
- **Data Requirements**:  
  - **Distances/Flight Times** (accounting for air traffic restrictions)  
  - **Airspace Classifications** (FAA guidelines, no-fly zones)  
- **Sources**:  
  - **Federal Aviation Administration (FAA)**: Airspace classification maps, Temporary Flight Restrictions (TFRs).  
  - **GIS Tools**: Calculate distances and potential flight paths between candidate nodes.

### 3.1.3 \( P \): Passenger Segments
- **Definition**: Each \(p \in P\) denotes a unique passenger type (e.g., business travelers, tourists, VIPs).  
- **Data Requirements**:  
  - **Demand Levels** (volume of each passenger type)  
  - **Willingness-to-Pay** or Fare Sensitivity  
  - **Travel Behavior** (peak vs. off-peak times, routes)  
- **Sources**:  
  - **Ride-Sharing Companies** (Uber, Lyft) for aggregated demand patterns.  
  - **Metropolitan Transportation Authority (MTA)**: Ridership by time of day.  
  - **Market Research** (surveys on eVTOL acceptance, paying capacity).

### 3.1.4 \( S \): Service (Aircraft) Types
- **Definition**: Each \(s \in S\) denotes a different eVTOL model or service configuration.  
- **Data Requirements**:  
  - **Technical Specs** (seats, range, charging/fueling needs)  
  - **Operating Costs** (maintenance, pilot salaries, energy usage)  
  - **Noise/Emissions Profiles** (for regulatory compliance)  
- **Sources**:  
  - **Manufacturer White Papers** (Joby, Archer, Volocopter, etc.).  
  - **NASA AAM Studies**: eVTOL performance assumptions.  
  - **Academic/Industry Research** on battery or hydrogen fuel cell technology.

### 3.1.5 \( T \): Time Periods
- **Definition**: Each \(t \in T\) could represent different times of day (peak/off-peak) or planning horizons (morning vs. evening).  
- **Data Requirements**:  
  - **Temporal Demand Variation** (rush hour vs. mid-day)  
  - **Operational Windows** (curfews, noise restrictions at night)  
- **Sources**:  
  - **NYC DOT Traffic Volume & Congestion Reports**: Time-of-day patterns.  
  - **Ride-Sharing Data**: Peak trip times.  
  - **FAA/Local Ordinances**: Possible nighttime restrictions.


## 3.2. Parameters

Here are **typical parameters** in a UAM Service Network Design model, along with **where to get** the corresponding data.

### 3.2.1 Revenue & Cost Parameters
- **$r_{p,a}$**: **Revenue** per unit of passenger type $p$ on arc $a$.  
  - *Data:* Fare willingness by passenger type, typical flight distance/time.  
  - *Sources:* Market surveys (business vs. tourist WTP), competitor helicopter taxi rates, ride-share route revenues.

- **$v_{p,a}$**: **Variable Cost** of transporting passenger type $p$ on arc $a$.  
  - *Data:* Energy/fuel costs, insurance per passenger, route-specific landing fees.  
  - *Sources:* eVTOL operating manuals, airport/heliport fee schedules, energy price data.

- **$c_{s,a}$**: **Service Operating Cost** for service $s$ on arc $a$.  
  - *Data:* Pilot labor, maintenance cost per flight hour, overhead for route licensing.  
  - *Sources:* eVTOL manufacturer O&M (operation & maintenance) guidelines, local regulations for route licensing.

- **$f_s$**: **Fixed Cost** for activating service $s$.  
  - *Data:* Lease/loan cost for eVTOL vehicles, insurance, pilot/crew salaries (if lumpsum).  
  - *Sources:* Manufacturer or leasing company quotes, pilot union rates.

### 3.2.2 Capacity & Demand
- **$u_{s,a}$**: **Capacity** of service $s$ on arc $a$ (max passengers/flight).  
  - *Data:* Seat count, weight restrictions, route distance constraints.  
  - *Sources:* eVTOL specifications, FAA flight manual.  

- **$d_{p,t}$**: **Demand** for passenger segment $p$ at time $t$.  
  - *Data:* Estimated number of travelers (by type, by time).  
  - *Sources:* MTA ridership data, ride-share trip data, tourism office stats.

- **$M$**: A **large constant** for frequency constraints.  
  - *Data:* Chosen artificially to exceed any realistic flight frequency in a period.  
  - *Sources:* Typically a modeling choice (e.g., set $M = 100$ or some upper bound on daily flights).

### 3.2.3 Feasibility & Accessibility
- **$\alpha_{p,a}$**: **Accessibility measure** (flight time, distance, or noise) for passenger $p$ on arc $a$.  
  - *Data:* Route distance, flight speed, noise thresholds.  
  - *Sources:* GIS distance calculations, eVTOL performance specs, local noise ordinances.

- **$\beta$**: **Threshold** (max distance/time/noise).  
  - *Data:* Maximum acceptable flight time or decibel limit.  
  - *Sources:* Local regulations, user surveys on acceptable travel time or noise exposure.

### 3.2.4 Budget & Coverage
- **$B$**: **Budget** available (optional).  
  - *Data:* Projected capital/operating funds from investors or agencies.  
  - *Sources:* Organizational financial statements, government grants, private investment documents.

- **$\kappa_s$**: **Minimum Service Coverage** requirement for service $s$.  
  - *Data:* Mandated minimum flights or coverage level per eVTOL type.  
  - *Sources:* Policy directives (city agreements), corporate brand/marketing commitments.

# 4. Why This Profit Maximization Model Is Well-Suited

- **Early-Stage Private Investment**  
  In a nascent AAM market, companies want to ensure financial viability. Achieving a positive margin is more critical than simply minimizing costs.

- **Selective Route Launch**  
  Cost minimization alone might lead to minimal coverage or suboptimal revenue. Profit maximization prompts operators to open only those routes/facilities that yield net-positive returns. Service $s$ in time $t$ is activated only if it yields net profit.


- **Scalability**  
  You can add or remove arcs/services as market demand evolves.

- **Strategic Insights**  
  Helps identify which vertiports, passenger segments, and flight frequencies are most profitable under capacity and travel constraints.
