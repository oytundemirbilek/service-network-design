#Service Network Design of Advanced Air Mobility Vertiports in NYC

# 1. Introduction

Advanced Air Mobility (AAM) is no longer just a concept—billions of dollars in global investment and numerous pilot projects show it is on the verge of commercial rollout. In February 2024, Archer Aviation announced successful flight tests of its ‘Midnight’ eVTOL with FAA oversight, moving the aircraft closer to commercial readiness. Meanwhile, Joby Aviation continues to secure key approvals, highlighting how eVTOL services could soon become operational in major cities.

Among these urban centers, New York City stands out as an ideal early launch market, thanks to its extreme congestion, existing heliport infrastructure, and high willingness-to-pay among residents, tourists, and business travelers. Yet to fully tap into NYC’s potential, a well-designed service network is crucial—operators must determine which routes to fly, how often to operate, and how to balance costs and demand. This case study focuses on those network design considerations, offering insights into route selection, scheduling, pricing strategies, and the infrastructure constraints that will shape the next generation of urban air travel.

# 2.2 Cost Model

The following text explains the **vertiport infrastructure** and cost model necessary for operating **Urban Air Mobility (UAM)**. The key idea is that a **distributed network of takeoff and landing sites** is required to maximize time savings and overall system throughput. In this scenario, an **initial fleet of 240 VTOL aircraft** is deployed across **New York City Area**, supported by **20 vertiports**, each capable of handling **12 VTOLs** simultaneously. The infrastructure is modeled after facilities similar to existing heliports in New York City. Below are the main points:



## 2.2.1 Initial Infrastructure Redevelopment Cost

- **\$30 million total for 20 vertiports**
  - Repurposing upper levels of parking garages or similar structures to serve as vertiports.
  - Includes structural modifications and charger installations at each vertiport.
  - **Chargers**:
    - 1 high-voltage charger per vertiport (approx. \$250,000 each).
    - 2 low-voltage chargers per vertiport (approx. \$10,000 each → \$20,000 total).
  - **Cost Amortization**: The \$30 million is spread out over 30 years.

<br>



## 2.2.2 Annual Vertiport Operating Cost

- **\$21,500 per VTOL per year**
  - With 240 VTOLs, totals \$5.16 million annually.
  - Covers lease fees, maintenance, security, personnel, and other operational expenses for the 83 main vertiports.

<br>



## 2.2.3 Operating Expenses for eVTOL Services



### 1) Pilot Cost

- **Pilot Cost**  
  - Fully burdened annual cost per pilot: **\$50,000**  
  - **1.5 pilots** needed per VTOL (to account for scheduling/rotations)
  - **\$50,000 * 1.5 pilots * 240 eVTOL = $18 million/per year**
  - Pilot training: Similar to existing **Part 135** commercial requirements, with annual recurrent training


- **Share of Direct Operating Cost**  
  - Overall piloting costs ≈  **36%** of baseline direct operating costs

<br>

### 2) Vehicle Maintenance Costs

- **Electric VTOL Advantage**  
  - Maintenance cost per flight hour assumed **~50% lower** than light helicopters  
  - Rationale: Elimination of complex rotor components, plus electric motors have fewer moving parts and can achieve ~10,000-hour Time Between Overhaul (TBO)

- **Maintenance Procedures**  
  - Daily visual inspections, 100-hour minor checks  
  - Major annual service to restore vehicle to original specifications

- **Share of Direct Operating Cost**  
  - Maintenance + labor ≈ **22%** of baseline direct operating costs

<br>

### 3) Indirect Operating Costs

- **Definition**  
  - Covers non-vehicle-specific expenses: credit card processing, registration/permit fees, insurance, and smaller miscellaneous fees

- **Commercial Aviation Comparison**  
  - Indirect costs can be ~50% above direct operating costs for major airlines (due to booking overhead, taxes, landing fees)

- **Assumption for eVTOL**  
  - Lower indirect overhead (use of private infrastructure, less reliance on large commercial airports)  
  - Currently modeled at **~12%** on top of direct operating costs in the baseline scenario

<br>

## 2.2.4 Summary of Cost Distribution

- **Initial Build-Out**: \$30 million for 20 vertiports (charging infrastructure included), amortized over 30 years. (**$50,000 per vertiport/year**)

- **Annual Opex**: \$21,500 per VTOL (or **\$5.16 million** for a 240-vehicle fleet).
- **Piloting/Avionics**: **$18 million/per year** (~36% of direct operating costs)
- **Maintenance**: ~22% of direct operating costs (thanks to simpler electric drivetrain)  
- **Indirect Costs**: ~12% added on top of total direct operating costs

## 2.3 Demand Estimation

The demand is estimated from taxi user datasets in NYC from Kaggle:

- **New York City Taxi Trip Duration Extended:**
This dataset, derived from 2016 NYC Yellow Cab records published by the TLC, contains attributes like pickup/dropoff times, locations, and newly added geographic/time-based fields. We use it to extract the demand for each route.

- **Uber Pickups in New York City:**
This dataset contains data on over 4.5 million Uber pickups in New York City from April to September 2014, and 14.3 million more Uber pickups from January to June 2015. There is only pick-up information available in this dataset but first-class and high-class customer pick-up locations can be used to weigh more on the potential demand since AAM is considered premium service at the early stage of the business.

# 3. Mathematical Model for NYC AAM Vertiport Service Network

We have built a profit maximization model because of the following reasons.

- **Early-Stage Private Investment**  
  Achieving a positive margin is more critical than simply minimizing costs.

- **Selective Route Launch**  
  The operator only opens routes/facilities that yield net-positive returns.

- **Scalability**  
  The arcs(routes) can be added or removed as market demand evolves.

- **Strategic Insights**  
  Pinpoints which vertiport(s), passenger segments, and route capacities are most profitable.


<br>

## 3.2

## 3.1 Sets & Indices

- $N$: The set of nodes  *(Index: $i \in N$, or $j \in N$ for distinct origin/destination)*

- $V$: The set of potential vertiports (nodes)  *(Index: $i \in N$, or $j \in N$ for distinct origin/destination)*

- $H$: The set of Hub

- $A$: The set of directed arcs $(i,j)$ representing feasible aerial routes   *(Index: $a \in A$)*



<br>

## 3.2 Parameters

**Revenue & Costs**
- $r_{a}$: Revenue per unit of passenger type $p$ on arc $a$  
- $c_{a}$: Operating cost for transporting passenger on arc $a$  
- $C_{n}$: Fixed cost for operating smaller vertiport

**Capacity & Demand**
- $d_{a}$: Demand for arc $a$
- $K_{h}$: Maximum capacity on hub $h$ (max vehicles/vertiport)
- $K_{n}$: Maximum capacity on vertiport $n$ (max vehicles/vertiport)  
- $S$: The maximum number of demand units(seats) a vehivle can accommodate

**Budget & Coverage**
- $B$: Available budget (optional constraint)  


<br>

## 3.3 Decision Variables

**Service Activation & Frequency**  
- $y_{n} \in \{0,1\}$: 1 if the smaller vertiport $n$ is open, else 0  
- $f_{a} \in \mathbb{N}_0$: Integer frequency of flights
- $u_{a} \in \mathbb{N}_0$: Unsatisfied demand

<br>

## 3.4 Objective Function: Maximize Profit

$$
\max \Bigg(
\underbrace{\sum_{a \in A} r_{a} \, (f_{a})}_{\text{Total Revenue}}
\;-\;
\underbrace{\Big[\sum_{i \in I}
C_{n} \cdot y_{n}
\;+\;
 \sum_{a \in A} c_{a} \,\cdot f_{a}
\Big]}_{\text{Total Cost}}
\Bigg)
$$

- **Total Revenue**: Summation of revenues from passenger flows on each arc.  
- **Total Cost**:  
  1. Fixed cost $C_n$ (if small vertiport n is activated).  
  2. Operating cost $c_{a}$ for passenger flows $f_{a}$.  
  

<br>

## 3.5 Constraints

**(1) Demand Satisfaction**

$$
 f_{a} \cdot S \;\ge\; d_{a} -  u_a,
\quad \forall a \in A.
$$

  - **Interpretation:** Frequency should be the smallest interger value that demand is satisfied.

<br>

**(2) Flow Conservation (if multi-leg routes exist)**

$$
 \sum_{j \in N} f_{a_{j,i}}
=
 \sum_{k \in N} f_{a_{i,k}},
\quad \forall n \in N
$$

   - **Interpretation:** The flow of passengers arriving at node $i$ must equal the flow departing node $i$. No passenger is created or lost en route.

<br>

**(3) Capacity Constraint**

$$
\sum_{a \in A} f_{a} \;\le\; c_n \cdot y_n
\quad \forall n \in N
$$

$$
\sum_{a \in A} f_{a} \;\le\; c_n \cdot y_n
\quad \forall n \in N
$$

   - **Interpretation:** If $y=0$ (the vertiport is not opened), then $z_{i}=0$. Otherwise, $z_{i}$ can be up to $u_{i}$.

<br>



**(7) Budget Constraint (Optional)**

$$
\sum_{i \in L}
f_{i} \cdot y_{i}
\;+\;
 \sum_{a \in A} v_{a} \, x_{a}
\;+\;
\sum_{i \in L} c \, z_{i}
\;\;\le\;\; B.
$$

   - **Interpretation:** The sum of fixed cost ($f \cdot y$) and variable flow costs ($v_{a} \cdot x_{a}$) must not exceed budget $B$.
